# 自动求导

```python
x = torch.arange(4.0)
x
```

```python
x.requires_grad_(True), x.grad  # 默认为None
```

```python
y = 2 * torch.dot(x, x)
y
```

```python
y.backward(retain_graph=True)
x.grad
```

```python
# pytorch 默认为累积梯度，因此需要清除
x.grad.zero_()
y = x.sum()
y.backward(retain_graph=True)
x.grad
```

```python
# 深度学习中我们计算的是批量中每个样本单独计算的偏导数之和
x.grad.zero_()
y = x * x
y.sum().backward()
x.grad  # 即每个样本的偏导数
```

## 分离计算

```python
x.grad.zero_()
y = x * x
u = y.detach()  # 去除y的梯度影响，视为常数
z = u * x
z.sum().backward()
x.grad  # z=u*x的导数为u，即y=x*x
```

```python
# 然后再计算y的导数
x.grad.zero_()
y.sum().backward()
x.grad  # y的导数为2*x
```
